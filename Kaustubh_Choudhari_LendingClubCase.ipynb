###### Handling Null Values

```python
#Finding percentage of null or missing values
null_value_perc = round(100*(df.isnull().sum()/len(df.index)), 2)
#Printing columns which have more than 0% missing values
null_value_perc[ null_value_perc > 0 ]
```

Removing columns which has more than 30% null values in it.


```python
df.drop(null_value_perc[ null_value_perc > 30 ].index, axis=1, inplace=True)
```

```python
df.nunique().sort_values().head(15)
```

**'tax_liens', 'delinq_amnt', 'chargeoff_within_12_mths', 'acc_now_delinq', 'application_type', 'policy_code', 'collections_12_mths_ex_med', 'initial_list_status', 'pymnt_plan'** columns have only one unique value in all the rows, which will not give any usefull info during our visualization. lets remove them for further analysis

```python
df.drop(['tax_liens', 'delinq_amnt', 'chargeoff_within_12_mths', 'acc_now_delinq', 'application_type', 'policy_code', 'collections_12_mths_ex_med', 'initial_list_status', 'pymnt_plan'], axis=1, inplace=True)
```

```python
#Finding number of unique values, printing high unique valued columns
df.nunique().sort_values(ascending=False).head(12)
```

**emp_title, and title** columns have only text values which are  unique values.

```python
df.drop(['emp_title','title'], axis = 1, inplace= True)
```

**id, url, and member_id** having all unique values which means any one of these can be used as primary key.
so we will using  **id** as our primary key and remove url and member id columns.

```python
df.drop(['url', 'member_id'], axis = 1, inplace=True)   #Here we can easily access the customer using his ID so we removed url & member_id
```

**total_rec_int, total_rec_prncp, total_rec_late_fee, recoveries, collection_recovery_fee, last_credit_pull_d, last_pymnt_d, out_prncp, out_prncp_inv** variables are valid for borrowers who already took loan. we are only interested only in loan application details so we will remove these columns.
**recoveries, collection_recovery_fee** columns are only valid for charged off loans. ( defaulters )
Will remove these columns as well.

```python
df.drop(['total_rec_int', 'total_rec_prncp', 'total_rec_late_fee', 'last_credit_pull_d', 'recoveries', 'collection_recovery_fee', 'last_pymnt_d'], axis=1, inplace=True)
df.drop(['out_prncp', 'out_prncp_inv'], axis=1, inplace=True)
```

```python
#Finding percentage of null or missing values
null_value_perc = round(100*(df.isnull().sum()/len(df.index)), 2)
null_value_perc[ null_value_perc > 0 ]
```

```python
df.dropna(subset=['emp_length', 'revol_util', 'pub_rec_bankruptcies'], inplace=True)

#Finding percentage of null or missing values
null_value_perc = round(100*(df.isnull().sum()/len(df.index)), 2)
null_value_perc[ null_value_perc > 0 ]
```

As in our analysis we are going to focus on the customers who fully charged the loan or are the defaulter ( i.e : charged_off) so we will be dropping 'CURRENT' from the loan_status column

```python
#Removing current loan status rows
df = df[df.loan_status != 'Current']
```

```python
#Checking unique values for term column    #Term column is basically the time period 
df.term.unique()
```

There is empty space at the start of values in term column. removing it.

```python
#Stripping empty space in values in term
df['term'] = df.term.str.strip()
df.term.unique()
```

```python
df.shape  
```

### Data Type Conversion

```python
df.dtypes
```

#### Derived Variable
###### Here we are converting our issue_d & earliest_cr_line column into seprate date, month & year column

```python
#issue_d column
df['issue_d_year'] = df.issue_d.dt.year
df['issue_d_month'] = df.issue_d.dt.strftime('%b')
df['issue_d_weekday'] = df.issue_d.dt.weekday
#data type conversion of year and weekday
df['issue_d_year'] = df['issue_d_year'].astype(object)
df['issue_d_weekday'] = df['issue_d_weekday'].astype(object)

#earliest_cr_line
df['earliest_cr_line_year'] = df.earliest_cr_line.dt.year
df['earliest_cr_line_month'] = df.earliest_cr_line.dt.strftime('%b')
#data type conversion of year and weekday
df['earliest_cr_line_year'] = df['earliest_cr_line_year'].astype(object)
```

Creating Approved Loan amount ratio which is a ratio of Funded Amount by investor to Requested Loan amount.

```python
df['approved_loan_amnt_ratio'] = round(df.funded_amnt_inv*100/df.loan_amnt,2)
```
